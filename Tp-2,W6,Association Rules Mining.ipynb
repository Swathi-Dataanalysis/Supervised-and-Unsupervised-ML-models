Association Rule Mining
Market Basket Analysis is one of the key techniques used by large retailers to uncover associations between items. It works by looking for combinations of items that occur together frequently in transactions. To put it another way, it allows retailers to identify relationships between the items that people buy.

Association Rules are widely used to analyze retail basket or transaction data and are intended to identify strong rules discovered in transaction data using measures of interestingness, based on the concept of strong rules.

Details of the dataset
The dataset has 38765 rows of the purchase orders of people from the grocery stores. These orders can be analysed and association rules can be generated using Market Basket Analysis by algorithms like Apriori Algorithm.

Apriori Algorithm
Apriori is an algorithm for frequent itemset mining and association rule learning over relational databases. It proceeds by identifying the frequent individual items in the database and extending them to larger and larger item sets as long as those item sets appear sufficiently often in the database. The frequent itemsets determined by Apriori can be used to determine association rules which highlight general trends in the database: this has applications in domains such as market basket analysis.

An example of Association Rules
Assume there are 100 customers
10 of them bought milk, 8 bought butter and 6 bought both of them.
bought milk => bought butter
support = P(Milk & Butter) = 6/100 = 0.06
confidence = support/P(Butter) = 0.06/0.08 = 0.75
lift = confidence/P(Milk) = 0.75/0.10 = 7.5

Note: this example is extremely small. In practice, a rule needs the support of several hundred transactions, before it can be considered statistically significant, and datasets often contain thousands or millions of transactions.

Some important terms:
Support: This says how popular an itemset is, as measured by the proportion of transactions in which an itemset appears.

Confidence: This says how likely item Y is purchased when item X is purchased, expressed as {X -> Y}. This is measured by the proportion of transactions with item X, in which item Y also appears.

Lift: This says how likely item Y is purchased when item X is purchased while controlling for how popular item Y is.

2. Explaination of the Apriori Algorithm
The Apriori algorithm is a classic data mining method used to extract frequent itemsets and derive association rules from a transactional database. It’s particularly popular in market basket analysis to identify products that frequently co-occur in transactions.

Intuition and Mathematical Representation

Intuition: The Apriori algorithm is based on the principle that a subset of a frequent itemset must also be frequent. Its name comes from the concept that the algorithm uses prior knowledge of frequent itemset properties. The algorithm iteratively extends frequent itemsets, one item at a time (a bottom-up approach), and uses a support threshold to prune non-frequent itemsets to minimize the number of candidate sets.
Mathematical Representation:
•	Support (S): The probability that a transaction randomly selected from the database contains all items in the itemset. Mathematically, it is defined as:
Support(X)=Number of transactions containing X/Total number of transactions
•	Confidence (C): Given two items X and Y, the confidence of the rule X→Y is the probability that a transaction containing X also contains Y. It is computed as:
Confidence(X→Y)=Support(X∪Y)/Support(X)
•	Lift (L): The lift of the rule X→Y measures how much more often X and Y occur together than expected if they were statistically independent. It is defined as:
Lift(X→Y)=Confidence(X→Y)/Support(Y)

Steps of the Apriori Algorithm
1.	Set a minimum support and confidence.
2.	Generate all itemsets of size 1.
3.	Calculate the support of these itemsets and prune those that are below the minimum support.
4.	Use the frequent itemsets of size nn to generate n+1 itemsets.
5.	Repeat until no further successful extensions are found.
6.	Use the frequent itemsets to generate association rules that meet the minimum confidence.

Assumptions
•	All subsets of a frequent itemset must also be frequent (Apriori property).
•	Items within a transaction or itemset are considered unordered.
•	The input format is a collection of sets (transactions), each transaction consisting of a set of items.

### Preprocess Data for Apriori Algorithm ###

Transaction Consolidation: Since each row in the dataset likely represents a single item in a transaction, we will need to consolidate rows based on transaction ID so that each transaction is represented as a list of items

Grouping Data: df.groupby(['Member_number', 'itemDescription'])['Date'].count()
This groups the data by both Member_number and itemDescription.
The ['Date'].count() portion counts the occurrences of each item for each member, effectively quantifying how many times each member bought each item. It assumes that if a member bought an item multiple times on different dates, it should count each occurrence.
Pivoting Data: .unstack()
The unstack() function is used to pivot the item descriptions into column headers. This turns the item descriptions into individual columns, with the counts as the cell values under each column for each member.
Resetting Index: .reset_index()
This resets the index of the DataFrame, turning the Member_number from an index into a regular column. This is useful if we need to perform operations that require Member_number as a column.
Handling Missing Values: .fillna(0)
Since not all members buy every item, there will be missing values in the DataFrame after unstacking. The fillna(0) method replaces all NaN values with 0, indicating that the member did not purchase that item.
Setting Index: .set_index('Member_number')
Finally, this sets Member_number back as the DataFrame index. This is useful for identification purposes and ensures that each row uniquely represents a member's purchases.

Analyze Item Frequency
The first step in EDA is often to get a sense of which items are purchased most frequently. This can be achieved by summing up each item column in the binary basket dataframe.

```python
# Using groupby function to create a basket df, which specifies if an item is bought by particular Member
# for all items and all Members
# this will denote the tranasaction happened in the particular day which needs to be fixed

basket = (df.groupby(['Member_number', 'itemDescription'])['Date'].count()
          .unstack().reset_index().fillna(0).set_index('Member_number'))
```

```python
# Convert counts to 1s and 0s
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
```

5: Building the Model with the Apriori Algorithm
Implementing the Apriori Algorithm:
Choose a minimum support threshold that is appropriate for the dataset. This value is often chosen after some trial and error; common starting points are in the 1-5% range.
Use the Apriori function from the mlxtend library to find frequent itemsets.
The algorithm will iterate through the baskets of items, finding all itemsets that meet the minimum support threshold.

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
```

This above code block will return the itemsets that have a support of at least 1%, which means they appear in at least 1% of all transactions.

6. Generate Association Rules(Running the Model):
With the frequent itemsets identified, the next step is to generate association rules from these itemsets.
Choose minimum thresholds for metrics such as confidence and lift to filter the most relevant rules.

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.1)  # Adjust the threshold as needed
```

10. Validation of the Model
Validation in the context of association rule mining is about ensuring that the rules generated are both statistically significant and make practical sense. 

Support measures the frequency with which the itemset appears in the dataset. It's important because it indicates the base popularity of the itemset regardless of other items. However, high support does not imply a strong association. It's also essential not to set the support so low that you get spurious associations.

Confidence measures the likelihood that the consequent is purchased when the antecedent is purchased. It's a direct measure of the strength of the association. However, a high confidence rule can be misleading if the consequent is also very common in general; this is where lift comes in.

Lift measures how much more often the antecedent and consequent of the rule are purchased together than expected if they were statistically independent. A lift greater than 1 indicates a positive relationship.

Conviction is another metric that can be thought of as a predictor of the strength of a rule. It compares the expected frequency that the antecedent appears without the consequent if they were dependent with the actual observed frequency of the antecedent without the consequent. A high conviction means the consequent is highly dependent on the antecedent.

Zhang's Metric is used to measure the reliability of an association rule, which can be positive or negative indicating the direction of the rule.

Advantages and Dis advantages of Association rule mining :
Advantages of using frequent item sets and association rule mining include:
Efficient discovery of patterns: Association rule mining algorithms are efficient at discovering patterns in large datasets, making them useful for tasks such as market basket analysis and recommendation systems.
Easy to interpret: The results of association rule mining are easy to understand and interpret, making it possible to explain the patterns found in the data.
Can be used in a wide range of applications: Association rule mining can be used in a wide range of applications such as retail, finance, and healthcare, which can help to improve decision-making and increase revenue.
Handling large datasets: These algorithms can handle large datasets with many items and transactions, which makes them suitable for big-data scenarios.

Disadvantages of using frequent item sets and association rule mining include:
Large number of generated rules: Association rule mining can generate a large number of rules, many of which may be irrelevant or uninteresting, which can make it difficult to identify the most important patterns.
Limited in detecting complex relationships: Association rule mining is limited in its ability to detect complex relationships between items, and it only considers the co-occurrence of items in the same transaction.
Can be computationally expensive: As the number of items and transactions increases, the number of candidate item sets also increases, which can make the algorithm computationally expensive.
Need to define the minimum support and confidence threshold: The minimum support and confidence threshold must be set before the association rule mining process, which can be difficult and requires a good understanding of the data.
